```python
suites=["Hearts","Diamonds","clubs","spades"]
Ranks=["one","two","three","four","five","six","seven","eight","nine","ten","jock","queen","king","ace"]
Values={"one":1,"two":2,"three":3,"four":4,"five":5,"six":6,"seven":7,"eight":8,"nine":9,"ten":10,"jock":11,"queen":12,"king":13,"ace":14}
```

```python
class Card:
    def __init__(self,suite,rank):
        self.suite=suite
        self.rank=rank
        self.value=Values[self.rank]
    def __str__(self):
        return self.rank+" of "+self.suite
```

```python
import random
class Deck:
    def __init__(self):
        self.all_cards=[]
        for i in suites:
            for j in Ranks:
                card=Card(i,j)
                self.all_cards.append(card)
    def __str__(self):
        return self.all_cards
    def display(self):
        print(self.all_cards)
    def shuffle(self):
        return random.shuffle(self.all_cards)#it shuffles all the cards for ex:l=[1,2,3] shuffle(l)=>[3,1,2]
    def deal_one(self):
        return self.all_cards.pop()
```

```python
class player:
    def __init__(self,name):
        self.name=name
        self.all_cards=[]
    def addcards(self,new_cards):
        if type(new_cards)==list:
            return self.all_cards.extend(new_cards)
        else:
            return self.all_cards.append(new_cards)
    def remove_one(self):
        return self.all_cards.pop(0)
    def __str__(self):
        return f'player {self.name} has {len(self.all_cards)} cards'
```

```python
player_one=player("one")
player_two=player("two")

new_deck=Deck()
new_deck.shuffle()



for x in range(26):
    player_one.addcards(new_deck.deal_one())
    player_two.addcards(new_deck.deal_one())
print(player_one)
print(player_two)
```

```python
game_on = True
round_no = 0

while game_on:
    round_no += 1
    print(f"Rounds-{round_no}")
    
    if len(player_one.all_cards) == 0:
        print("player one has no more cards....")
        print("player two won!!!")
        game_on = False
        break
        
    if len(player_two.all_cards) == 0:
        print("player two has no more cards....")
        print("player one won!!!")
        game_on = False
        break

    player_one_cards = []
    player_two_cards = []
    
    player_one_card = player_one.remove_one()
    player_two_card = player_two.remove_one()
    
    player_one_cards.append(player_one_card)
    player_two_cards.append(player_two_card)
  
    at_war = True
    
    while at_war:
        print("length-1", len(player_one_cards), player_one_cards[0])
        print("length-2", len(player_two_cards), player_two_cards[0])
        
        print(player_one_cards[-1].value)
        print(player_two_cards[-1].value)
        
        if player_one_cards[-1].value < player_two_cards[-1].value:
            player_two.addcards(player_one_cards)
            player_two.addcards(player_two_cards)  
            
            at_war = False
            
        elif player_one_cards[-1].value > player_two_cards[-1].value:
            player_one.addcards(player_one_cards)
            player_one.addcards(player_two_cards)   
            
            at_war = False
            
        else:
            print('WAR!')
            print(len(player_one.all_cards))
            print(len(player_two.all_cards))
            if len(player_one.all_cards) < 3:
                print("player one unable to declare war!!!")
                print("player two wins!!!")
                game_on = False
                break
            elif len(player_two.all_cards) < 3:
                print("player two unable to declare war!!!")
                print("player one wins!!!")
                game_on = False
                break
            else:
                for i in range(3):
                    player_one_cards.append(player_one.remove_one())
                    player_two_cards.append(player_two.remove_one())
```
